```python
import newspaper
```

```python
def extract_article_text(url):
    # Download and parse the article
    article = newspaper.Article(url)
    article.download()
    article.parse()
    
    # Return the article's full text
    return article.text
```

```python
url = "https://www.w3schools.com/js/js_strings.asp"
article_text = extract_article_text(url)
print(article_text)
```
